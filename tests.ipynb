```python
import pandas as pd
from datetime import datetime,timedelta
import numpy as np
df = pd.read_csv("dags/predictive_maintenance.csv",parse_dates=["date_data"])
```

```python
from datetime import datetime, timedelta

# Função para definir a data com base no valor de UDI
def calcular_data(udi):
    if udi <= 2000:
        return datetime.today().date() - timedelta(days=4)
    elif udi <= 4000:
        return datetime.today().date() - timedelta(days=8)
    elif udi <= 8000:
        return datetime.today().date() - timedelta(days=12)
    else:
        return datetime.today().date() - timedelta(days=16)

# Aplicando a função na coluna "UDI"
df["date_data"] = df["UDI"].apply(calcular_data)
df["date_data"] = pd.to_datetime(df["date_data"])
```

```python
df.to_csv("dags/predictive_maintenance.csv",index = False)
```
